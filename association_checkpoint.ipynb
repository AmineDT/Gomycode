```python
import pandas as pd

# Define the dataset
dataset = [
    ['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
    ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']
]

# Preprocess the dataset
unique_items = set([item for transaction in dataset for item in transaction])
preprocessed_dataset = [list(set(transaction)) for transaction in dataset]

# Create a DataFrame for visualization
df = pd.DataFrame(preprocessed_dataset)
df.columns = ['Item{}'.format(i+1) for i in range(len(df.columns))]

# Display the preprocessed dataset
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the dataset into one-hot encoded format
one_hot_encoded = pd.get_dummies(df, prefix='', prefix_sep='')

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(one_hot_encoded, cmap='YlGnBu', cbar=False)
plt.title('Item Combination Heatmap')
plt.xlabel('Items')
plt.ylabel('Transactions')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Apply the Apriori algorithm to generate frequent itemsets
frequent_itemsets = apriori(one_hot_encoded, min_support=0.2, use_colnames=True)
frequent_itemsets
```

```python
# Extract association rules from the frequent itemsets
association_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)

# Display the association rules
print(association_rules[['antecedents', 'consequents', 'support', 'confidence']])
```
