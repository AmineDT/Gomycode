```python
import pandas as pd
import numpy as np

titanic_df = pd.read_csv("titanic.csv")
titanic_df
```

```python
titanic_df.isna().sum()
```

```python
titanic_df.drop(["cabin", "boat", "body", "home.dest"], inplace=True, axis=1)
titanic_df.isna().sum()
```

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].median())
titanic_df.isna().sum()
```

```python
titanic_df = titanic_df.dropna().reset_index(drop=True)
titanic_df.isna().sum()
```

```python
titanic_df
```

```python
titanic_df.describe()
```

```python
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()

        df = df[(df[col] <= mean+(n_std*sd))]
        
        
    return df
```

```python
titanic_dfc = remove_outliers(titanic_df, ["age", "sibsp", "parch", "fare"], 3)
titanic_dfc
```

```python
diff_df = titanic_df[~titanic_df.isin(titanic_dfc)].dropna()
diff_df
```

```python
print("Clean columns stats : \n")
titanic_dfc_stat = titanic_dfc.describe()
print(titanic_dfc_stat.iloc[1:2 , [3, 4, 5]])

print("\n\n\n\n")

print("Removed columns stats : \n")
diff_df_stat = diff_df.describe()
print(diff_df_stat.iloc[1:2 , [3, 4, 5]])
```

```python
from sklearn.preprocessing import LabelEncoder

# Instantiate a LabelEncoder object
le = LabelEncoder()
```

```python
titanic_dfc['sex_encoded'] = le.fit_transform(titanic_dfc['sex'])
titanic_dfc['embarked_encoded'] = le.fit_transform(titanic_dfc['embarked'])
titanic_dfc = titanic_dfc.reset_index(drop=True)
titanic_dfc['Family'] = titanic_dfc.parch + titanic_dfc.sibsp

titanic_dfc
```

```python
titanic_dfc["Title"] = titanic_dfc.name.apply(lambda name: name.split(",")[1].split(".")[0])
titanic_dfc
```

```python
titanic_dfc["Title"].unique()
```

```python
titanic_dfc['title_encoded'] = le.fit_transform(titanic_dfc['Title'])
```

```python
titanic_dfc.drop(["name", "sex", "embarked", "Title"], axis = 1, inplace = True)
titanic_dfc
```
