```python
import pandas as pd
import numpy as np

titanic_df = pd.read_csv("titanic.csv")
titanic_df
```

```python
titanic_df.isna().sum()
```

```python
titanic_df.drop(["cabin", "boat", "body", "home.dest"], inplace=True, axis=1)
titanic_df.isna().sum()
```

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].median())
titanic_df.isna().sum()
```

```python
titanic_df = titanic_df.dropna().reset_index(drop=True)
titanic_df.isna().sum()
```

```python
titanic_df
```

```python
titanic_df.describe()
```

```python
def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()

        df = df[(df[col] <= mean+(n_std*sd))]
        
        
    return df
```

```python
titanic_dfc = remove_outliers(titanic_df, ["age", "sibsp", "parch", "fare"], 3)
titanic_dfc
```

```python
df_compare = pd.concat([titanic_df, titanic_dfc])
df_compare.shape
```

```python
df_compare.drop_duplicates(inplace=True)
df_compare.shape
```
