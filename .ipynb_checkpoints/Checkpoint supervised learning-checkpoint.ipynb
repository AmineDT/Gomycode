```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('kc_house_data.csv')

# Explore the dataset
# Check for missing values
data.isnull().sum()

```

```python
# Check data types
data.dtypes
```

```python
# Calculate descriptive statistics
data.describe()
```

```python
# Create a correlation matrix
corr_matrix = data.corr()

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, annot=True, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
ax.set_title('Correlation Matrix')

plt.show()
```

```python
price_correlations = corr_matrix['price'].sort_values(ascending=False)

# Print the correlations in descending order
print(price_correlations)
```

# We see have ordered the features by correlation to take the most meaningful ones

# I - Linear regression checkpoint

```python
# Select important features
selected_features = ['sqft_living', 'grade', 'sqft_above', 'sqft_living15', 'bathrooms', 'view', 'view', 'bedrooms']

# Split the dataset into a training set and a testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data[selected_features], data['price'], test_size=0.3, random_state=42)

# Apply linear regression to the training set
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Plot the linear regression line
plt.figure(figsize=(10, 8))
plt.scatter(X_train['sqft_living'], y_train)
plt.plot(X_train['sqft_living'], lin_reg.predict(X_train), color='red')
plt.xlabel('Square footage of living area')
plt.ylabel('Price')
plt.title('Linear Regression')
plt.show()
```

# There are some outliers

```python
# Measure the performance of linear regression using metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
y_pred = lin_reg.predict(X_test)
linear_reg_mse = mean_squared_error(y_test, y_pred)
linear_reg_mae = mean_absolute_error(y_test, y_pred)
linear_reg_r2 = r2_score(y_test, y_pred)
```

```python
# Apply multiple linear regression and measure its performance
multi_reg = LinearRegression()
multi_reg.fit(X_train, y_train)
y_pred_multi = multi_reg.predict(X_test)
multi_reg_mse = mean_squared_error(y_test, y_pred_multi)
multi_reg_mae = mean_absolute_error(y_test, y_pred_multi)
multi_reg_r2 = r2_score(y_test, y_pred_multi)
```

```python
# Apply polynomial regression and measure its performance
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.transform(X_test)
poly_reg.fit(X_train_poly, y_train)
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_train_poly, y_train)
y_pred_poly = poly_lin_reg.predict(X_test_poly)
poly_reg_mse = mean_squared_error(y_test, y_pred_poly)
poly_reg_mae = mean_absolute_error(y_test, y_pred_poly)
poly_reg_r2 = r2_score(y_test, y_pred_poly)
```

```python
# Create a dataframe to store the results of the three models
model_results = pd.DataFrame({'Linear Regression': [linear_reg_mse, linear_reg_mae, linear_reg_r2],
                               'Multiple Linear Regression': [multi_reg_mse, multi_reg_mae, multi_reg_r2],
                               'Polynomial Regression': [poly_reg_mse, poly_reg_mae, poly_reg_r2]},
                               index=['Mean Squared Error', 'Mean Absolute Error', 'R-Squared'])

model_results
```

# Comparison :
## Polynomial regression has the lowest meaned squared error and mean absolute error, and the highest R-Squared. It's the best model
